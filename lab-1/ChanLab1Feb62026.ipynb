```python
# Cody Chan
# Lab 1 Assignment - Aquiring Text data using API

# Install and Call required packages
!pip install requests
import requests
import os
import json

# API Key from Serpapi
SERPAPI_API_KEY = "" # Remove for posting
# Set parameters for youtube transcript
params = {
    "api_key": SERPAPI_API_KEY,
    "engine": "youtube_video_transcript",
    "v": "ToTh7_CQUGU", # KCOC Winter News
    "type": "asr"
}

# Search youtube video on Serpapi
search = requests.get("https://serpapi.com/search", params=params)
response = search.json() # Output the structured data from video
print(json.dumps(response, indent=2))

# Include time stamps in the transcript
for result in response.get("transcript", []): # Loop through transcript file
  snippet = result.get("snippet") # # Get values from result
  start_time_text = result.get("start_time_text")

  # Print time stamp with SnippetL
  print(f"SnippetL: {snippet}")
  print(f"Start Time: {start_time_text}")
  print("-" * 10)

# Create new file and store the transcript
with open('transcript.txt', 'w') as f: # Open new file and write it
  for result in response.get("transcript", []): # Loop through transcript
    snippet = result.get("snippet") # extract contents
    if snippet:
      f.write(f"{snippet}\n") # write text on file and create a newline
print("Results saved to transcript.txt")
```
